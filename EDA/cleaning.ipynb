```python
import pandas as pd
import numpy as np
import string
import re
import nltk
from nltk.corpus import stopwords
import inflect
```

```python
tweets = pd.read_csv("./data/usernames/LostAndFoundLON.csv", index_col=0)
# tweets.dropna(inplace=True)
len(tweets)
```

```python
string.punctuation
def remove_punctuation(text):
    words_wo_punct= text.translate(str.maketrans('', '', string.punctuation))
    return words_wo_punct.lower()
tweets['text'] = tweets['text'].apply(lambda x: remove_punctuation(x))
tweets.head()
```

```python
def tokenize(text):
    split=re.split("\W+",text) 
    return split
tweets['text']=tweets['text'].apply(lambda x: tokenize(x.lower()))
tweets.head()
```

```python
nltk.download('stopwords')
stop_words = stopwords.words('english')
def remove_stopwords(text):
    words_wo_stop = [word for word in text if not word in stop_words]
    return words_wo_stop
tweets['text'] = tweets['text'].apply(lambda x: remove_stopwords(x))
tweets.head()
```

```python
p = inflect.engine()
def singular(text):
    word_list = []
    for word in text:
        if word != '':
            if p.singular_noun(word) != False:
                word_list.append(p.singular_noun(word))
            else:
                word_list.append(word)
    return word_list
tweets['text'] = tweets['text'].apply(lambda x: singular(x))
tweets.head()
```

```python
tweets.to_csv('LostAndFoundLON_clean.csv')
```
